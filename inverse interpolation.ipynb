```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
func = lambda x:  x * np.exp(-x) * 10 - 2 #x ** 5 - 5 * x + 3
x = np.linspace(0.5, 3, 20)
h = x[1] - x[0]
y = func(x)
```

```python
import matplotlib.pyplot as plt
plt.plot(x, y)
plt.show()
```

```python
def get_deltas(points_y):
    deltas = np.array([[0 for _ in range(len(points_y))] for _ in range(len(points_y))], dtype='float32')
    deltas[:,0] = points_y
    for j in range(1, deltas.shape[1]):
        for i in range(deltas.shape[0] - j):
            deltas[i, j] = deltas[i+1, j-1] - deltas[i, j-1]
    return deltas

deltas = get_deltas(y)
pd.DataFrame(get_deltas(y))
```

```python
def interpolation(point, xs, deltas, h):
    u = []
    u = list(filter(lambda x: x <= 0, xs - point))
    if u != []:
        index = np.argmax(np.array(u))
    else:
        index = 0
    x0 = xs[index] # point = x0 + t * h так как узлы равноотстоящие друг от друга
    y0 = deltas[index, 0]
    t = np.abs((x0 - point) / h)
    p = y0
    m = t
    f = 1
    ind = index
    sign = 1
    temp = ind
    
    i = 1
    while temp != 0:
        p += (deltas[index, i]) * m / f
        m *= (t - sign)
        sign += 1
        i += 1
        temp -= 1
        f *= (i + 1)
    return p
```

```python
def inverse_interpolation(xs, deltas, h, e = 1e-1):
    for i in range(len(deltas[:,0])-1):
        if (deltas[i,0]*deltas[i+1,0]<0):
            index = i
            break
    x0 = xs[index]
    temp = []
    point = x0
    temp.append(point)
    t = interpolation(point, xs, deltas, h)
    temp.append(t)
    while (np.abs(temp[-2]-temp[-1])> 1e-6):
        f = (interpolation(x0 + e, x, deltas, h)-interpolation(x0 - e, x, deltas, h))/(2 * e)
        t -= interpolation(t, xs, deltas, h)/float(f)
        temp.append(t)
    point = t
    return point, temp

point = inverse_interpolation(x, deltas, h)[0]
temp = inverse_interpolation(x, deltas, h)[1]
print(point, func(point))
temp
```
