```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
%config InlineBackend.figure_format='retina'
```

```python
def mandelbrot(pmin, pmax, ppoints, qmin, qmax, qpoints, 
               max_iterations=200, infinity_border=10000):
    image = np.zeros((ppoints, qpoints))
    p, q = np.mgrid[pmin:pmax:(ppoints*1j), qmin:qmax:(qpoints*1j)]
    c = p + 1j*q
    z = np.zeros_like(c)
    for k in range(max_iterations):
        z = z**2 + c
        mask = (np.abs(z) > infinity_border) & (image == 0)
        image[mask] = k
        z[mask] = np.nan
    return -image.T


p_center, q_center = -0.5, 0
scalefactor = 2

pmin_ = p_center - scalefactor/2 
qmin_ = q_center - scalefactor/2 
pmax_ = p_center + scalefactor/2 
qmax_ = q_center + scalefactor/2 

plt.figure(figsize=(10, 10))
image = mandelbrot(pmin_, pmax_, 1000, qmin_, qmax_, 1000, 500)
plt.xticks([])
plt.yticks([])
plt.imshow(image, cmap='rainbow', interpolation='none')
```
